1.  Liste todos os códigos e nomes de funções da subfunção que possui o maior número de funções.

```python
df = pd.read_sql_query("SELECT COD_FUNCAO, NOME_FUNCAO "+
                       "FROM (DETALHAMENTO_FUNCAO NATURAL JOIN FUNCAO) "+
                       "WHERE COD_SUBFUNCAO = (SELECT COD_SUBFUNCAO "+
                                               "FROM DETALHAMENTO_FUNCAO "+
                                               "GROUP BY COD_SUBFUNCAO "+
                                               "ORDER BY COUNT(COD_FUNCAO) DESC LIMIT 1);", conn)
df
```

4. Liste o mês e ano, modalidade, programa de governo e valor de todas as despesas do órgão Ministério da Educação no primeiro semestre de 2018.


```python
df = pd.read_sql_query("SELECT MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, NOME_PROGRAMA_GOVERNO, VALOR_DESPESA "+
                       "FROM (DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO) "+
                       "WHERE COD_DESPESA IN (SELECT COD_DESPESA "+
                                           "FROM (DESPESA_PUBLICA NATURAL JOIN ORGAO) "+
                                           "WHERE NOME_ORGAO = 'Ministério da Educação' "+
                                           "AND ANO_DESPESA = 2018 AND MES_DESPESA <= 6);", conn)
df
```

5. Liste o código e nome dos órgãos que possuíram os 5 maiores valores de despesas totais nas funções de educação e saúde durante o ano de 2018.

```python
df = pd.read_sql_query("SELECT COD_ORGAO, NOME_ORGAO "+
                       "FROM (ORGAO NATURAL JOIN DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN FUNCAO) "+
                       "WHERE (NOME_FUNCAO = 'Educação' OR NOME_FUNCAO = 'Saúde') AND ANO_DESPESA = 2018 "+
                       "GROUP BY COD_ORGAO "+
                       "ORDER BY SUM(VALOR_DESPESA) DESC LIMIT 5;", conn)
df
```

6. Liste todos os nomes de funções e subfunções das despesas públicas dos últimos 5 meses de 2018 e seus respectivos valores totais, ordenando de forma decrescente pelo valor.

```python
df = pd.read_sql_query("SELECT NOME_FUNCAO, NOME_SUBFUNCAO, SUM(VALOR_DESPESA) AS DESPESA_TOTAL "+
                       "FROM (DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN FUNCAO NATURAL JOIN SUBFUNCAO) "+
                        "WHERE ANO_DESPESA = 2018 AND MES_DESPESA >= 8 "+
                        "GROUP BY NOME_FUNCAO, NOME_SUBFUNCAO "+
                        "ORDER BY SUM(VALOR_DESPESA) DESC;", conn)
df
```

7. Liste os programas de governo e a quantidade de órgãos governamentais distintos envolvidos nas despesas públicas desses programas. Nesta consulta, desconsidere valores de despesa negativos ou iguais a zero no período.


```python
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, COUNT(DISTINCT COD_ORGAO) AS QTDE_ORGAOS "+
                       "FROM (PROGRAMA_GOVERNO NATURAL JOIN DESPESA_PUBLICA) "+
                       "WHERE VALOR_DESPESA > 0 "+
                       "GROUP BY COD_PROGRAMA_GOVERNO;", conn)
df
```

8. Liste os nomes das subfunções das despesas públicas envolvidas para cada órgão governamental, apenas nos casos em que o valor total das despesas excedeu 10 mil reais. Sua consulta deve retornar o nome do órgão, a subfunção da despesa, a quantidade de entradas de despesas públicas e o valor total dessas despesas.


```python
df = pd.read_sql_query("SELECT NOME_ORGAO, NOME_SUBFUNCAO, COUNT (VALOR_DESPESA) AS QTDE_DESPESAS, SUM(VALOR_DESPESA) AS TOTAL_DESPESAS "+ 
                       "FROM (SUBFUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN DESPESA_PUBLICA NATURAL JOIN ORGAO) "+
                       "GROUP BY NOME_ORGAO, NOME_SUBFUNCAO HAVING SUM(VALOR_DESPESA) > 10000;", conn)
df
```

12. πNOME_FUNCAO,VALOR_DESPESA(FUNCAO ⋈ DETALHAMENTO_FUNCAO ⋈ DESPESA_PUBLICA)


```python
df = pd.read_sql_query("SELECT DISTINCT NOME_FUNCAO, VALOR_DESPESA "+
                       "FROM (FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN DESPESA_PUBLICA);", conn)
df
```

13.  πMODALIDADE_DESPESA,VALOR_DESPESA,NOME_ORGAO(ORGAO ⋈ σMODALIDADE_DESPESA='Reserva de Contingência'(DESPESA_PUBLICA))

```python
df = pd.read_sql_query("SELECT  DISTINCT MODALIDADE_DESPESA, VALOR_DESPESA, NOME_ORGAO "+
                       "FROM (ORGAO NATURAL JOIN (SELECT * FROM DESPESA_PUBLICA "+
                       "WHERE MODALIDADE_DESPESA = 'Reserva de Contingência'));", conn)
df
```

14. πVALOR_DESPESA,NOME_PROGRAMA_GOVERNO(DESPESA_PUBLICA ⋈ σNOME_PROGRAMA_GOVERNO='ENEM'(PROGRAMA_GOVERNO))

```python
df = pd.read_sql_query("SELECT DISTINCT VALOR_DESPESA, NOME_PROGRAMA_GOVERNO "+
                       "FROM (DESPESA_PUBLICA NATURAL JOIN (SELECT * FROM PROGRAMA_GOVERNO "+
                       "WHERE NOME_PROGRAMA_GOVERNO = 'ENEM'));", conn)
df
```

15.  πNOME_PROGRAMA_GOVERNO,QTD_ORGAOS_SUBORDINADOS,VALOR_DESPESA(DESPESA_PUBLICA ⋈ σNOME_ORGAO='Ministério da Educação'(ORGAO) ⋈ PROGRAMA GOVERNO)


```python
df = pd.read_sql_query("SELECT DISTINCT NOME_PROGRAMA_GOVERNO, QTD_ORGAOS_SUBORDINADOS, VALOR_DESPESA "+
                       "FROM (DESPESA_PUBLICA NATURAL JOIN (SELECT * FROM ORGAO "+
                       "WHERE NOME_ORGAO = 'Ministério da Educação') "+
                       "NATURAL JOIN PROGRAMA_GOVERNO);", conn)
df
```
